```python
import pandas as pd

# Sample DataFrame with categorical variables
data = {
    'Category': ['A', 'B', 'C', 'A', 'C'],
    'Color': ['Red', 'Blue', 'Green', 'Green', 'Red'],
    'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# One-Hot Encoding
one_hot_encoded = pd.get_dummies(df, columns=['Category', 'Color', 'Size'], drop_first=True)
print("One-Hot Encoded:")
print(one_hot_encoded)
```

```python

# Label Encoding
label_encoded = df.copy()
label_encoded['Category'] = label_encoded['Category'].astype('category').cat.codes
label_encoded['Color'] = label_encoded['Color'].astype('category').cat.codes
label_encoded['Size'] = label_encoded['Size'].astype('category').cat.codes
print("\nLabel Encoded:")
print(label_encoded)
```

```python
# Mean Encoding for 'Category' column
mean_encoded = df.copy()
category_mean = mean_encoded.groupby('Category')['Category'].transform('count').astype(float)
mean_encoded['Category'] = category_mean
print("\nMean Encoded:")
print(mean_encoded)
```

```python
# Frequency Encoding
frequency_encoded = df.copy()
frequency_encoded['Category'] = frequency_encoded.groupby('Category')['Category'].transform('count')
frequency_encoded['Color'] = frequency_encoded.groupby('Color')['Color'].transform('count')
frequency_encoded['Size'] = frequency_encoded.groupby('Size')['Size'].transform('count')
print("\nFrequency Encoded:")
print(frequency_encoded)
```
